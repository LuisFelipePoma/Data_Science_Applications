## Imagen de prueba

```python
import cv2
import matplotlib.pyplot as plt
```

```python
X = cv2.imread('/content/peru_sf.jpeg')
plt.figure(figsize=(12,30))
plt.imshow(cv2.cvtColor(X, cv2.COLOR_BGR2RGB))
plt.title('ingUC.jpg')
plt.show()
```

## Detección de caras

```python
from mtcnn.mtcnn import MTCNN
detector = MTCNN()
```

```python
face_locations = detector.detect_faces(X)
```

## Despliegue de resultados

```python
plt.figure(figsize=(12,30))
plt.imshow(cv2.cvtColor(X, cv2.COLOR_BGR2RGB))
plt.title('ingUC.jpg')

n = len(face_locations)
for i in range(n):
  bb = face_locations[i]['box']
  x1 = bb[0]
  x2 = bb[0]+bb[2]
  y1 = bb[1]
  y2 = bb[1]+bb[3]
  x = [x1,x1,x2,x2,x1]
  y = [y1,y2,y2,y1,y1]
  plt.plot(x,y,c='red')

plt.show()
```
